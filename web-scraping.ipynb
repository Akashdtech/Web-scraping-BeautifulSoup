```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://draft.dev/learn/technical-blogs/'

web_page = requests.get(url)
web_page
```

```python
web_page.content
```

```python
Parser = BeautifulSoup(web_page.content, 'html.parser')
print(Parser)
```

```python
print(Parser.prettify())
```

```python
page_title = Parser.title.string
page_title
```

```python
scraped_best_blogs = Parser.find_all('h3', class_ = 'post-title')
scraped_best_blogs
```

```python
best_blogs = []
for blog in scraped_best_blogs:
    best_blogs.append(blog.text)

best_blogs
```

```python
scraped_description = Parser.find_all('p', class_ = 'post-summary')
scraped_description
```

```python
blogs_description = []
for description in scraped_description:
    blogs_description.append(description.text)

blogs_description
```

```python
data = pd.DataFrame()
data['Best Technical Blogs'] = best_blogs
data['Description'] = blogs_description
data
```

```python
data.to_csv('Best_Technical_Blogs.csv', index = False)
```
